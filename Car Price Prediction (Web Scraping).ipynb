```python
# importing libraries
import selenium
import pandas as pd
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException
```

```python
# connecting to webdriver
driver = webdriver.Chrome(r"C:\Users\Nitish\Downloads\Compressed\chromedriver.exe")
time.sleep(2)
```

```python
# send get request to the webpage server to get the source code of the page
url = 'https://www.cardekho.com/'
driver.get(url) 
time.sleep(4)
```

```python
# Empty Lists
a1=[]
a2=[]
a3=[]
a4=[]
a5=[]
a6=[]
a7=[]
a8=[]
a9=[]
```

```python
# scraping model names
try:
    for i in driver.find_elements_by_xpath("//div[@class='gsc_col-xs-7 carsName']/a"):
        a1.append(i.text)
except NoSuchElementException:
    a1.append("No detail")
    
# scraping variant names    
try:
    for i in driver.find_elements_by_xpath("//div[@class='gsc_col-xs-7 carsName']/div[1]"):
        a2.append(i.text)
except NoSuchElementException:
    a2.append("No detail")

# scraping transmission tags
try:
    for i in driver.find_elements_by_xpath("//div[@class='truncate dotlist']/span[3]"):
        a3.append(i.text)
except NoSuchElementException:
    a3.append("No detail")
```

```python
# extracting urls 
urls=[]
for links in driver.find_elements_by_xpath('//div[@class="gsc_col-xs-7 carsName"]/a'):
    urls.append(links.get_attribute('href'))
```

```python
# to check the number of urls extracted
len(urls)
```

```python
# Iterating through the URL list and accessing each link to extract data    
for url in urls:
    driver.get(url)
    time.sleep(4)
    
    # scraping make year & mileage tags 
    try:
        for i in driver.find_elements_by_xpath("//ul[@class='gsc_row detailsList']/li[1]/div"):
            a4.append(i.text)
    except NoSuchElementException:
        a4.append("No detail")
     
    # scraping fuel & max power tags
    try:
        for i in driver.find_elements_by_xpath("//ul[@class='gsc_row detailsList']/li[3]/div"):
            a5.append(i.text)
    except NoSuchElementException:
        a5.append("No detail")
    
    # scraping kilometers driven & torque tags
    try:
        for i in driver.find_elements_by_xpath("//ul[@class='gsc_row detailsList']/li[4]/div"):
            a6.append(i.text)
    except NoSuchElementException:
        a6.append("No detail")
    
    # scraping engine displacement tags
    try:
        for i in driver.find_elements_by_xpath("//ul[@class='gsc_row detailsList']/li[5]/div"):
            a7.append(i.text)
    except NoSuchElementException:
        a7.append("No detail")
    
    # scraping owner tags 
    try:
        for i in driver.find_elements_by_xpath("//ul[@class='gsc_row detailsList']/li[6]/div"):
            a8.append(i.text)
    except NoSuchElementException:
        a8.append("No detail")
    
    # scraping price tags
    try:
        for i in driver.find_elements_by_xpath("//div[@class='priceSection']/span[2]"):
            a9.append(i.text)
    except NoSuchElementException:
        a9.append("No detail")
```

```python
# stripping to exact values
lists1=[]
for i in range(0,len(a4),2):
    lists1.append(a4[i].strip().replace('Make Year\n',''))
    
lists2=[]
for i in range(1,len(a4),2):
    lists2.append(a4[i].strip().replace('Mileage\n',''))

lists3=[]
for i in range(0,len(a5),2):
    lists3.append(a5[i].strip().replace('Fuel\n',''))

lists4=[]
for i in range(1,len(a5),2):
    lists4.append(a5[i].strip().replace('Max Power\n',''))

lists5=[]
for i in range(0,len(a6),2):
    lists5.append(a6[i].strip().replace('KMs Driven\n',''))

lists6=[]
for i in range(1,len(a6),2):
    lists6.append(a6[i].strip().replace('Torque\n',''))
    
lists7=[]
for i in range(0,len(a7),2):
    lists7.append(a7[i].strip().replace('Engine Displacement\n',''))

lists8=[]
for i in range(0,len(a8),2):
    lists8.append(a8[i].strip().replace('No Of Owner(s)\n',''))
```

```python
cars1=pd.DataFrame({})            # making dataframe
cars1['Model']= a1
cars1['variant']= a2
cars1['mfg']= lists1
cars1['km_drvn']= lists5
cars1['fuel']= lists3
cars1['mileage']= lists2
cars1['engine']= lists7
cars1['max_pwr']= lists4
cars1['torque']= lists6
cars1['owner']= lists8
cars1['transmission']= a3
cars1['Price']= a9
cars1
```
